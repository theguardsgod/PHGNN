```python
import os
import shutil

source_dir = "/home/ubuntu22/dataset/adni/processed_mri_mat_ori"
target_dir = "/home/ubuntu22/dataset/adni/adni2/AD"
moved_dir = "/home/ubuntu22/dataset/adni/mat_adni_1mm/adni2/AD"
```

```python
# 获取AD目录下的所有文件名（去除后缀）
ad_filenames = [os.path.splitext(file)[0] for file in os.listdir(target_dir)]

# 遍历AD目录下的文件名
for ad_filename in ad_filenames:
    # 构建源文件路径
    source_file = os.path.join(source_dir, ad_filename + ".mat")
    # 构建目标文件路径
    target_file = os.path.join(moved_dir, ad_filename + ".mat")
    # 判断源文件是否存在，存在则复制到目标文件夹
    if os.path.exists(source_file):
        shutil.copyfile(source_file, target_file)
        print(f"Copied {source_file} to {target_file}")
    else:
        print(f"File {ad_filename}.mat does not exist in the source directory.")
```
