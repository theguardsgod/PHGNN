```python
# Load the MRI data from the npy file
mri_data = np.load('/home/ubuntu22/dataset/adni/adni1/AD/002_S_0619.npy')
# Save the data as a NIfTI file
nifti_img = nib.Nifti1Image(mri_data, affine=np.eye(4))  # Assuming identity affine for simplicity
nib.save(nifti_img, '/home/ubuntu22/dataset/adni/adni1/AD/002_S_0619.nii.gz')
# Assuming mri_data is a 3D array representing MRI volumes, you can visualize a slice:
slice_num = 50  # Adjust this to visualize different slices
slice_data = mri_data[:, :, slice_num]

# Plot the slice using Matplotlib
plt.imshow(slice_data, cmap='gray')
plt.title('MRI Slice')
plt.colorbar()
plt.show()
```

```python
import os
import shutil

source_dir = "/home/ubuntu22/dataset/adni/processed_mri_mat_ori"
target_dir = "/home/ubuntu22/dataset/adni/adni2/AD"
moved_dir = "/home/ubuntu22/dataset/adni/mat_adni_1mm/adni2/AD"
```

```python
# AD
ad_filenames = [os.path.splitext(file)[0] for file in os.listdir(target_dir)]

# AD
for ad_filename in ad_filenames:
    # 
    source_file = os.path.join(source_dir, ad_filename + ".mat")
    # 
    target_file = os.path.join(moved_dir, ad_filename + ".mat")
    # 
    if os.path.exists(source_file):
        shutil.copyfile(source_file, target_file)
        print(f"Copied {source_file} to {target_file}")
    else:
        print(f"File {ad_filename}.mat does not exist in the source directory.")
```
